## Exploratory Data Analysis (EDA)

This notebook performs visual exploration on the cleaned food nutrition dataset.
The analysis includes:
- Univariate analysis
- Bivariate analysis
- Multivariate analysis

The goal is to understand data distributions, relationships, and patterns
that will guide feature engineering and model selection.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/clean_food_data.csv')
df.shape
```

```python
df.columns
```

```python
numeric_cols = [
    'Calories','Protein','Fat','Carbs','Sugar','Fiber',
    'Sodium','Cholesterol','Glycemic_Index',
    'Water_Content','Serving_Size'
]

categorical_cols = ['Meal_Type', 'Preparation_Method']
```

```python
# Univariate Analysis of Numeric Features for Calories
plt.figure(figsize=(6,4))
sns.histplot(df['Calories'], kde=True)
plt.title("Calories Distribution")
plt.show()

plt.figure(figsize=(6,2))
sns.boxplot(x=df['Calories'])
plt.title("Calories Outliers")
plt.show()
```

- The calories distribution is right-skewed with a long tail toward higher values.
- Most food items fall within a moderate calorie range, while a smaller number
of items have very high calorie values, indicating a heavy-tailed distribution.


```python
plt.figure(figsize=(8,2))
sns.boxplot(x=df['Calories'])
plt.title("Calories Outliers")
plt.xlabel("Calories")
plt.show()
```

- The boxplot shows several high-end outliers in calorie values.
- These outliers likely correspond to calorie-dense food items and are
considered valid real-world observations rather than data errors.

```python
df['Calories'].kurtosis()
```

- The kurtosis value indicates a heavy-tailed distribution,
confirming the presence of extreme calorie values.


### Calories  EDA Insights

- Calories exhibit a right-skewed distribution with a heavy tail.
- Most food items are clustered in the lower to mid-calorie range,
while a small number of foods contain extremely high calories.
- Outliers are retained as they represent realistic, calorie-dense foods.

# BIVARIATE ANALYSIS

- Numerical vs Numerical (Calories vs Fat)

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Fat', y='Calories', data=df, alpha=0.5)
plt.title("Fat vs Calories")
plt.xlabel("Fat")
plt.ylabel("Calories")
plt.show() 
```

- A strong positive relationship is observed between fat and calories, indicating that foods with higher fat content tend to have higher calorie values.
- This suggests fat is a key contributor to overall caloric content.

Categorical vs Numerical (Preparation Method vs Fat)

```python
#Categorical vs Numerical
plt.figure(figsize=(8,4))
sns.boxplot(x='Preparation_Method', y='Fat', data=df)
plt.title("Fat Content by Preparation Method")
plt.show()
```

- Fat content varies across preparation methods, with fried foods showing higher fat levels compared to raw or baked items.
- This highlights the influence of preparation method on nutritional composition.


Target vs Numerical (Calories by Food Category)

```python
#Target vs Numerical (Sample Only)
plt.figure(figsize=(10,4))
sns.boxplot(x='Food_Name', y='Calories', data=df)
plt.xticks(rotation=90)
plt.title("Calories by Food Category")
plt.show()
```

- Different food categories exhibit distinct calorie distributions, indicating that nutritional features can help differentiate food types.


# MULTIVARIATE ANALYSIS

```python
# Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols].corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Between Nutritional Features")
plt.show()
```

- Strong correlations are observed among calorie-related features such as fat, sugar, and carbohydrates. 
- These relationships suggest potential feature redundancy and indicate the need for feature scaling and regularization during model training.


## EDA Summary and Modeling Implications

- Nutritional features show skewed distributions with meaningful outliers.
- Fat, sugar, and sodium exhibit strong relationships with calorie content.
- Food categories differ significantly in nutritional profiles.
- Preparation methods influence fat and calorie levels.
- Correlation analysis reveals strong dependencies among features,
  indicating the need for feature scaling and robust models.
- The dataset is now ready for feature engineering and model training.
